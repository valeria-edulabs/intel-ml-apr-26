# Topic 08: Statistical Visualizations

This notebook focuses on understanding data distributions and variability using Histograms and Box Plots.

```python
import pandas as pd
import plotly.express as px
import os

DATA_PATH = "../data/sensors_sample.csv"
```

## 1. Histograms
Visualizing the distribution of sensor values to check for process stability and 

```python
df = pd.read_csv(DATA_PATH, index_col=0)

fig_hist = px.histogram(
    df, 
    x='val3', 
    color='chamber', 
    nbins=50,
    title='Distribution of Sensor val3 by Chamber',
    marginal='box' # Adds a box plot on top
)
fig_hist.show()
```

## 2. Box Plots
Comparing sensor variability across tools to identify potential maintenance needs.

```python
fig_box = px.box(
    df, 
    x='tool', 
    y='val4', 
    title='Sensor val4 Variability per Tool',
    points='outliers'
)
fig_box.show()
```
