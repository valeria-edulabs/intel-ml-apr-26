# Topic 05: Plotly Basics

This notebook introduces interactive visualization using Plotly Express. Interactive plots allow engineers to zoom in on specific time intervals and inspect individual data points on hover.

```python
import pandas as pd
import plotly.express as px
import os

DATA_PATH = "../data/sensors_sample.csv"
```

## 1. Preparing Data for Plotting
We'll filter for a single tool to keep the visualization clear.

```python
df = pd.read_csv(DATA_PATH, index_col=0)
df['ts'] = pd.to_datetime(df['ts'])

target_tool = df['tool'].iloc[0]
df_tool = df[df['tool'] == target_tool].sort_values('ts')
df_tool.head()
```

## 2. Interactive Line Chart
Creating a trend line for sensor `val3`.

```python
fig = px.line(
    df_tool, 
    x='ts', 
    y='val3', 
    title=f'Sensor val3 Trend for Tool {target_tool}',
    labels={'ts': 'Timestamp', 'val3': 'Sensor Reading'}
)

fig.show()
```
