# Topic 07: Advanced Plotly Features

This notebook covers high-level Plotly features like faceting (subplots) and time-series range sliders.

```python
import pandas as pd
import plotly.express as px
import os

DATA_PATH = "../data/sensors_sample.csv"
```

## 1. Faceting (Automatic Subplots)
Comparing multiple tools side-by-side using the `facet_col` parameter.

```python
df = pd.read_csv(DATA_PATH, index_col=0)
df['ts'] = pd.to_datetime(df['ts'])

# Select a few tools for the demo
selected_tools = df['tool'].unique()[:3]
df_filtered = df[df['tool'].isin(selected_tools)].sort_values('ts')

fig = px.line(
    df_filtered,
    x='ts',
    y='val3',
    facet_col='tool',
    color='chamber',
    title='Sensor val3: Comparison Across Tools'
)

# Add a range slider for time navigation
fig.update_xaxes(rangeslider_visible=True)

fig.show()
```
