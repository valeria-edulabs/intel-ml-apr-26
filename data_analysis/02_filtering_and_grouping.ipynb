# Topic 02: Filtering & Grouping

This notebook covers advanced data selection techniques and statistical grouping. These skills are essential for identifying process deviations and comparing hardware performance.

```python
import pandas as pd
import os

DATA_PATH = "../data/sensors_sample.csv"
df = pd.read_csv(DATA_PATH, index_col=0)
```

```python
df[df['val3'] > 5.5]
```

## 1. Value-Based Filtering (Anomaly Detection)
Identifying records where sensor values exceed typical operating limits.

```python
anomalies = df[df['val3'] > 5.5]
print(f"Detected {len(anomalies)} anomalies in sensor val3.")
anomalies.head()
```

## 2. Complex Logical Filtering
Using `|` (OR), `&` (AND), and `~` (NOT) to perform precise data selection.

```python
# Data for Chambers 1 or 5, but excluding Tool 11
complex_filter = df[((df['chamber'] == 1) | (df['chamber'] == 5)) & (df['tool'] != 11)]
complex_filter.head()
```

## 3. Statistical Grouping
Aggregating data by tool to compare performance benchmarks.

```python
tool_performance = df.groupby('tool')[['val3', 'val4']].mean()
tool_performance.sort_values('val3', ascending=False).head()
```

## 4. Pivot Tables
Creating a cross-tabulation of mean values across two dimensions: Tool and Chamber.

```python
pivot = df.pivot_table(values='val3', index='tool', columns='chamber', aggfunc='mean')
pivot.head()
```

## 5. Multi-level Aggregation
Using `.agg()` to calculate multiple metrics at once.

```python
chamber_stats = df.groupby(['tool', 'chamber'])['val3'].agg(['mean', 'std', 'count'])
chamber_stats.head(10)
```
