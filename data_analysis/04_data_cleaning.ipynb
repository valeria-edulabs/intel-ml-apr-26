# Topic 04: Data Cleaning & Imputation

Missing data can bias results or cause model failures. This notebook explores both basic and sophisticated methods for handling missing values in sensor datasets.

```python
import pandas as pd
import os

DATA_PATH = "../data/sensors_sample.csv"
df = pd.read_csv(DATA_PATH, index_col=0)
df['ts'] = pd.to_datetime(df['ts'])
```

## 1. Initial Missing Data Audit
Counting nulls across all sensors.

```python
df.isnull().sum()
```

## 2. Basic Imputation: Forward Fill
Propagating the last known value forward. This is valid for sensors that change state infrequently.

```python
df_ffill = df.sort_values('ts').ffill()
df_ffill.isnull().sum()
```

## 3. Sophisticated Imputation: Linear Interpolation
Creating a mathematical line between the point before and after the gap. This produces a smoother, often more realistic transition for analog sensors.

```python
df_interp = df.sort_values('ts').interpolate(method='linear')
df_interp.isnull().sum()
```

## 4. Advanced Imputation: Group-Specific Means
Instead of using a global average (which might be way off for different tools), we fill missing values using the average of that specific Tool/Chamber combination.

```python
# Using transform to broadcast the group mean back to the original rows
df['val5_cleaned'] = df.groupby(['tool', 'chamber'])['val5'].transform(lambda x: x.fillna(x.mean()))

print(f"Missing values in val5_cleaned: {df['val5_cleaned'].isnull().sum()}")
df[['tool', 'chamber', 'val5', 'val5_cleaned']].head(15)
```
