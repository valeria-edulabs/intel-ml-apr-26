# Topic 01: Loading & Initial Inspection

This notebook demonstrates how to load the sensor dataset and perform an initial inspection of the data structure, types, and summary statistics.

```python
import pandas as pd
import os

# In a notebook, the working directory is usually the folder containing the notebook.
# We'll set the path to the data relative to this folder.
DATA_PATH = "../data/sensors_sample.csv"
```

## 1. Loading the Data
We use `index_col=0` because the first column in our CSV is an unnamed index.

```python
df = pd.read_csv(DATA_PATH, index_col=0)

print("First 5 rows of the dataset:")
df.head(7)
```

## 2. Dataset Information
Checking data types and non-null counts.

```python
print("Dataset Info:")
df.info()
```

## 3. Statistical Summary
A quick overview of the distribution of numeric columns.

```python
print("Statistical Summary:")
df.describe()
```

## 4. Unique Identifiers
Identifying the tools and chambers represented in this sample.

```python
print(f"Unique Tools: {df['tool'].unique()}")
print(f"Unique Chambers: {df['chamber'].unique()}")
```
