# Topic 06: Multivariate Analysis

This notebook demonstrates how to compare multiple sensors simultaneously to identify relationships and correlations.

```python
import pandas as pd
import plotly.express as px
import os

DATA_PATH = "../data/sensors_sample.csv"
```

## 1. Scatter Plots for Correlation
Visualizing the relationship between `val3` and `val4`. We use sampling to ensure the interactive chart remains smooth with large datasets.

```python
df = pd.read_csv(DATA_PATH, index_col=0)
df_sample = df.sample(2000)

fig = px.scatter(
    df_sample,
    x='val3',
    y='val4',
    color='chamber',
    hover_data=['tool', 'ts'],
    title='Correlation: val3 vs val4 (Sampled Data)',
    template='plotly_dark'
)

fig.show()
```
